```
import time

start = time.time()

def fib(n):
  if n <= 1:
    return n
  else:
    return fib(n-2) + fib(n-1)

n = int(input("Enter the number to computer fibonacci number :"))
print("Fibonacci nth number is ", fib(n))

end = time.time()
print("Execution time:", end-start)
```
